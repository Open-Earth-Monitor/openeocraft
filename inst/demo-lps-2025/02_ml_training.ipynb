import openeo 

```python
connection = openeo.connect(url="http://127.0.0.1:8000")
```

```python
connection.authenticate_basic("brian", "123456")
```

```python
tempcnn_model_init = connection.mlm_class_tempcnn(
     optimizer="adam",
     learning_rate=0.0005,
     seed=42,
     verbose=True
)
```

```python
tempcnn_model = tempcnn_model_init.fit(
     training_set= "https://github.com/e-sensing/sitsdata/raw/main/data/samples_deforestation_rondonia.rds",
     target="label"
)
```

```python
result = tempcnn_model.save_ml_model(name ="tempcnn_rondonia", tasks=["classification"],
                           options={"mlm:accelerator":"macos-arm", "mlm:framework":"Torch"}
                        )
```

```python
job = result.create_job(
    title="Save TempCNN model (Rondonia 2022)",
    description="Train and persist the TempCNN model in the backend"
)
```

```python
job.start_and_wait()
results = job.get_results()
```
